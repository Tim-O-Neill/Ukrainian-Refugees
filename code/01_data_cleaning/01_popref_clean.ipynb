```python
rf = df.groupby(['Year', 'Country of asylum', 'Country of asylum (ISO)'])["Refugees under UNHCR's mandate", 'Asylum-seekers',
       'IDPs of concern to UNHCR', 'Stateless persons', 'Others of concern'].sum().reset_index()
```

```python
rf = rf.sort_values(by = ['Country of asylum', 'Year'])
```
